```python
#Import OpenCV
import os, cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
img_ref = cv2.imread("img3_ref.jpg", cv2.IMREAD_GRAYSCALE)
img_def = cv2.imread("img3_def.jpg", cv2.IMREAD_GRAYSCALE)
```

## 題目
### 1.  Training set 是兩兩一組的IC圖形, 一組有defect, 一組是沒defect的reference組
### 2.  總共有4種情況 
#### 1) 兩張都沒 defect 
#### 2) 有defect code1
#### 3) defect code2
#### 4) defect code3

### 3. 目標, train 一個 model 來分辨未來兩兩一組(referece組疑似defect組) defect 種類

```python
#初步看一下圖片長寬
plt.imshow(img_def, cmap='gray')
plt.show()
print('img_def shape' + str(img_def.shape))
plt.imshow(img_ref, cmap='gray')
plt.show()
print('img_ref shape' + str(img_ref.shape))
```

```python
#Crop裁切
#因為目前看到的 defect 都是在圖片正中間
#如果不是在正中間裁切方法可能要改
template = img_def[332:,34:][:-332,:-34]
#img_ref = img_ref[5-2:,13-3:][:-5+2,:-13+3]
plt.imshow(template, cmap='gray')
plt.show()
print(template.shape)
plt.imshow(img_ref, cmap='gray')
plt.show()
print(img_ref.shape)
```

```python
# 使用openCV的template matching 找到 reference 沒有defect 圖片相應的位置
# All the 6 methods for comparison in a list
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

w, h = template.shape[::-1]

for meth in methods:
    img = img_ref.copy()
    method = eval(meth)

    # Apply template Matching
    res = cv2.matchTemplate(img,template,method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

    # If the method is TM_SQDIFF or TM_SQDIFF_NORMED, take minimum
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)

    cv2.rectangle(img,top_left, bottom_right, 255, 2)

    plt.subplot(121),plt.imshow(res,cmap = 'gray')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(img,cmap = 'gray')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.suptitle(meth)

    plt.show()
```

```python
img_sub = np.copy(img_ref[top_left[1]:bottom_right[1],top_left[0]:bottom_right[0]])
```

```python
print(img_sub.shape)
```

```python
#兩張圖片相減
#因為顏色不同, 或是光源些許變化, 或是template matching的誤差, 所以會有許多雜訊
#需要更好的方法做 noise filtering
img_diff = np.abs(img_sub-template)
plt.imshow(img_diff, cmap='gray')
plt.show()

plt.hist(img_diff.ravel(),256,[0,256])
plt.show()
```

```python
img_diff[img_diff>155] = 0
img_diff[img_diff<80] = 0
```

```python
plt.imshow(img_diff, cmap='gray')
plt.show()
```

```python
# Data augmentation 增加樣本數 
# 旋轉, 翻轉
for i in range(0,4):
    rows, cols = img_diff.shape
    M = cv2.getRotationMatrix2D((cols/2, rows/2), i*90, 1)
    img_diff_r = cv2.warpAffine(img_diff, M, (cols, rows))
    img_diff_f = cv2.flip(img_diff_r, 1)
    plt.imshow(img_diff_r, cmap='gray')
    plt.show()
    plt.imshow(img_diff_f, cmap='gray')
    plt.show()
```
